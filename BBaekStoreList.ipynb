# 한번에 해보자!

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select 

# 데이터 담을 리스트
global shop, area, addr, tel 
shop = []
area = []
addr = []
tel = []

# 크롬웹드라이버 세팅
URL = "https://paikdabang.com/store/"
wd = webdriver.Chrome(r'./webDriver/chromedriver.exe')
wd.get(URL)

# 페이지네이션 반복문에 쓸 매장 정보 크롤링 함수
def StoreInfo():
    WebDriverWait(wd,2).until(EC.presence_of_element_located((By.ID,"pagination")))
    WebDriverWait(wd,2).until(EC.presence_of_element_located((By.ID,"store_list")))
    WebDriverWait(wd,2).until(EC.presence_of_element_located((By.CLASS_NAME,"page-numbers")))
    
    soup = BeautifulSoup(wd.page_source, 'html.parser')
    table = soup.find('tbody')
    store_shop = table.select('tr.tr_list > td.td_shop')
    
    # 데이터 에쁘게 뽑아서 각 리스트에 담기
    for i in store_shop: # 매장이름
        shop.append(i.string)
    store_area = table.select('tr.tr_list > td.td_area')
    for i in store_area: # 매장지역
        area.append(i.string)
    store_addr = table.select('tr.tr_list > td.td_addr')
    for i in store_addr: # 매장주소
        addr.append(i.string)
    store_tel = table.select('tr.tr_list > td.td_tel')
    for i in store_tel: # 매장전화
        tel.append(i.string)
    
    return
    
    
# 셀레니움으로 페이지 넘기며 매장 정보 로딩 후 크롤링하기
StoreInfo()
for i in range(2, 123):
    if i <= 4:
        WebDriverWait(wd,2).until(EC.presence_of_element_located((By.ID,"pagination")))
        page = wd.find_element(By.XPATH, "//*[@id=\"pagination\"]/li["+ str(i) + "]")
        page.click()
        StoreInfo()
        print(i, '&')
        time.sleep(1)
    else: 
        WebDriverWait(wd,2).until(EC.presence_of_element_located((By.ID,"pagination")))
        page = wd.find_element(By.XPATH, "//*[@id=\"pagination\"]/li[6]")
        page.click()
        StoreInfo()
        print(i, '*')
        time.sleep(2)
        
store = {'shop': shop,
        'area': area,
        'addr': addr,
         'tel': tel}
store_df = pd.DataFrame(store)
store_df.to_csv('./BBaekStore.csv', encoding='utf-8', mode='w', index=True)
```
