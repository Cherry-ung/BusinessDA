```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys

global item
item = []
global itemInfo
itemInfo = []
global rank, title, auth, pub, pubdate
rank = []
title = []
auth = []
pub = []
pubdate = []
bestseller = {'rank':rank,
             'book_title':title,
             'author':auth,
             'publisher':pub,
             'published_date':pubdate}

wd = webdriver.Chrome(r'./webDriver/chromedriver.exe')
wd.get("https://product.kyobobook.co.kr/category/KOR/32")
time.sleep(1)

# 베스트셀러 탭 누르기
wd.find_element("xpath", "//*[@id=\"ui-id-43\"]").click() # <- selenium 업뎃되면서 method 바뀌었넹
time.sleep(2)

for i in range(34): # i=0이 첫페이지
    if i >= 1 :
        next_page_btn = wd.find_element("xpath", "//*[@id=\"bestTopPagi\"]/button[2]")
        next_page_is = next_page_btn.is_enabled()
        print(i, next_page_is)
        next_page_btn.send_keys(Keys.ENTER)
        time.sleep(2)
    soup2 = BeautifulSoup(wd.page_source, 'html.parser') # type(soup): bs4.BeautifulSoup
    # WebDriverWait(wd,2).until(EC.presence_of_element_located((By.ID,"homeTabBest")))
    # WebDriverWait(wd,2).until(EC.presence_of_element_located((By.CLASS_NAME,"switch_prod_wrap")))
    bestList = soup2.select("div.switch_prod_wrap > ol.prod_list > li.prod_item")
    for i in bestList:
        # print(i.prettify())
        item.append(i.prettify())
print(item)

def str_to_bs4(x):
    html_soup = BeautifulSoup(x, 'html.parser')
    html_tags = html_soup.find_all('span')
    return html_tags

for i in item:
    itemInfo.append(str_to_bs4(i))

for i in range(0,len(itemInfo)):
    rank.append(itemInfo[i][4].string.strip()) 
    title.append(itemInfo[i][5].string.strip())
    auth.append(itemInfo[i][6].get_text().split()[0].strip())
    pub.append(itemInfo[i][6].get_text().strip().split('·')[1].strip())
    pubdate.append(itemInfo[i][6].get_text().strip().split('·')[2].strip('출시').strip())

print(len(rank), len(title), len(auth), len(pub), len(pubdate))
```
